```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
df = pd.read_csv('https://raw.githubusercontent.com/KryssyCo/DS-Unit-1-Sprint-5-Data-Storytelling-Blog-Post/master/national_afcars_trends_08thru17.updated2.csv')
print(df.shape)
df.head()
```

```
df.isnull().sum()
```

```
df.dtypes
```

```
df['Adopted'] = df['Adopted'].str.replace(',','').astype(int)
df['Entered'] = df['Entered'].str.replace(',','').astype(int)
df['Exited'] = df['Exited'].str.replace(',','').astype(int)
df['In Care on Sept 30th'] = df['In Care on Sept 30th'].str.replace(',','').astype(int)
df['Served'] = df['Served'].str.replace(',','').astype(int)
df['Parental Rights Terminated'] = df['Parental Rights Terminated'].str.replace(',','').astype(int)
df['Waiting for Adoption'] = df['Waiting for Adoption'].str.replace(',','').astype(int)
```

```
df.dtypes
```

```
sns.set(style='ticks', color_codes=True)
graph = sns.pairplot(df)
```
